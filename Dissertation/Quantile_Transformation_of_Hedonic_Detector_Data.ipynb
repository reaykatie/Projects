# Transforming the data to obtain normality in each variable

### Dependent Variables
The dependent variables were transformed using the QuantileTransformer (sometimes called the rank-transformer). A rank transformation smooths out unusual distributions. The median of the input to the transformation becomes the mean of the output, centered at 0.

### Independent Variables

All independent variables had their minimum value shifted to 0, and all but `PANAS_positive` were transformed with 

$$f: x\rightarrow \ln(x+1)$$

NB: `PANAS_positive` had low skew (-0.008) and thus did not need to be transformed.

Lastly, all independent variables were normalized, in that the mean was shifted to 0, and the standard deviation to 1.

### Other
Nothing was done to `Gender` and `Age`

```python
from sklearn.preprocessing import StandardScaler, QuantileTransformer, MinMaxScaler
sc = StandardScaler()
qt = QuantileTransformer(output_distribution='normal')
```

```python
ind_vars = ['PANAS_positive','PANAS_negative','ANX_score','DEPRESSION_score','STRESS_score']
dp_vars = [col for col in data.columns if col.startswith('MEAN')]
other = ['Age', 'Gender']
# Exception: PANAS_positive

ind_data_vals = (data[ind_vars]-data[ind_vars].min()).applymap(lambda x: np.log(x+1))
ind_data_transformed = pd.DataFrame(sc.fit_transform(ind_data_vals), columns=ind_vars)


dp_data_vals = data[dp_vars]
dp_data_transformed = pd.DataFrame(qt.fit_transform(dp_data_vals), columns=dp_vars)

data_transformed = pd.concat([data[other], ind_data_transformed, dp_data_transformed], axis=1)

data_transformed['PANAS_positive'] = (data['PANAS_positive'] - data['PANAS_positive'].mean())/data['PANAS_positive'].std()
data_transformed.head()
```

# Controlling for Anxiety

For each dependent variable in our analysis, a linear regression was run using `ANX_score` as the predictor variable. 


$$\text{Dependent Variable}_i = b_i*\text{ANX_score} + \text{intercept}_i + \epsilon_i$$

All subsequent linear regressions were run on the residuals ($\epsilon_i$).

### Computation of residuals

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()

for col in dp_vars:

    lr.fit(data_transformed[['ANX_score']], data_transformed[col])
    data_transformed[col+'_RESIDUAL'] = data_transformed[col] - lr.predict(data_transformed[['ANX_score']])

residual_variables = list(map(lambda x: x+'_RESIDUAL', dp_vars))
data_transformed[residual_variables].head()
```

### Confirming the residuals follow a normal distribution

```python
data_transformed[residual_variables].plot(kind='hist', figsize=(20,6), subplots=True, layout=(2,4))
plt.tight_layout()
```

Further confirmation with q-q plots

```python
import probscale

fig, ax = plt.subplots(ncols=4, nrows=2)
fig.set_size_inches(20, 10)

for i, var in enumerate(dp_vars):
    probscale.probplot(data_transformed[var+'_RESIDUAL'], ax=ax[i//4, i%4], plottype='qq',
                             problabel='Standard Normal Quantiles', 
                             datalabel=var,
                             scatter_kws=dict(marker='.', linestyle='none'))

    sns.despine()
plt.tight_layout()
```

# $\beta$, F, p and $R^2$ values

```python
i_vars = ['DEPRESSION_score','PANAS_positive']
```

```python
for i in residual_variables:
    for j in i_vars:
        lr_tmp = LinearRegression()
        lr_tmp.fit(data_transformed[[i]], data_transformed[j])
        eq = f'{str(i).ljust(30)}= {str(lr_tmp.coef_[0].round(3)).rjust(7)} * {j.ljust(18)} + {lr_tmp.intercept_.round(3)}'
        
        [f],[p] = f_regression(data_transformed[[i]], data_transformed[j])
        ds = f'{eq} | F = {str(f.round(2)).ljust(6)}| p = {str(p.round(4)).ljust(6)} | R\u00b2 = {lr_tmp.score(data_transformed[[i]], data_transformed[j]).round(3)}'
        print()
        print(ds)
```
